```python
from ultralytics import YOLO

def evaluate_model(model):
    model = YOLO(f"trained_models/yolo/{model}/weights/best.pt")

    properties = {
        "save_json": True,
        "plots": True,
        "imgsz": 640,
        "batch": 16,
        "conf": 0.50,
        "iou": 0.6,
        "device": "0",
        "save_txt": True,
        "save_conf": True,
        "save_crop": True
    }

    nominal_results = model.val(
        data="./datasets/LARD-real-nominal/dataset.yaml",
        **properties,
    )
    
    edge_cases_results = model.val(
        data="./datasets/LARD-real-edge-cases/dataset.yaml",
        **properties,
    )

    return nominal_results, edge_cases_results
```

```python
from ultralytics import YOLO

def train_model(variant, dataset, name):
    model = YOLO(f"yolo11{variant}-seg.pt")
    model.train(
        data=f"./datasets/{dataset}/dataset.yaml",
        epochs=100,
        patience=10,
        project="trained_models/yolo",
        name=f"{variant}_{name}",
    )

train_model("m", "LARD-yolo", "lard")
train_model("m", "p_BaseImages", "gard-p_BaseImages")
train_model("m", "p_VariantImages", "gard-p_VariantImages")
train_model("m", "p_VariantImagesWithOcclusion", "gard-p_VariantImagesWithOcclusion")

train_model("s", "LARD-yolo", "lard")
train_model("s", "p_BaseImages", "gard-p_BaseImages")
train_model("s", "p_VariantImages", "gard-p_VariantImages")
train_model("s", "p_VariantImagesWithOcclusion", "gard-p_VariantImagesWithOcclusion")
```

```python
import os
from ultralytics import YOLO

def evaluate_all_models():
    """
    {
      "model_folder_name": {
        "nominal": <results>,
        "edge_cases": <results>
      }, 
      ...
    }
    """
    
    models_base_path = "trained_models/yolo"
    properties = {
        "save_json": True,
        "plots": True,
        "imgsz": 640,
        "batch": 16,
        "conf": 0.50,
        "iou": 0.6,
        "device": "0",
        "save_txt": True,
        "save_conf": True,
        "save_crop": True
    }
    
    results_dict = {}
    
    for model_dir in os.listdir(models_base_path):
        model_path = os.path.join(models_base_path, model_dir)

        print(f"TESTING {model_dir}")
        
        best_weights_path = os.path.join(model_path, "weights", "best.pt")
        if not os.path.isdir(model_path) or not os.path.isfile(best_weights_path):
            continue  # Skip anything that isn't a valid model directory
        
        model = YOLO(best_weights_path)

        nominal_results = model.val(
            data="./datasets/LARD-real-nominal/dataset.yaml",
            **properties,
        )
        edge_cases_results = model.val(
            data="./datasets/LARD-real-edge-cases/dataset.yaml",
            **properties,
        )
        
        results_dict[model_dir] = {
            "nominal": nominal_results,
            "edge_cases": edge_cases_results
        }

    return results_dict
```

```python
results = evaluate_all_models()
```

```python
def extract_map_info(results):
    summary = {}

    for model_name, evaluations in results.items():
        model_summary = {}

        for scenario, result in evaluations.items():
            try:
                det_metrics = result.box  # detection metrics
                seg_metrics = result.seg  # segmentation metrics

                model_summary[scenario] = {
                    "detection/mAP50": det_metrics.map50,
                    "detection/mAP75": det_metrics.map75,
                    "detection/mAP50-95": det_metrics.map,
                    "segmentation/mAP50": seg_metrics.map50,
                    "segmentation/mAP75": seg_metrics.map75,
                    "segmentation/mAP50-95": seg_metrics.map,
                }

            except Exception as e:
                print(f"Error extracting metrics for model {model_name}, scenario {scenario}: {e}")
                model_summary[scenario] = "Failed to extract"

        summary[model_name] = model_summary

    return summary

map_summary = extract_map_info(results)

import json
print(json.dumps(map_summary, indent=2))
```
